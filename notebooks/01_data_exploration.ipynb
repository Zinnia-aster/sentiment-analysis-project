```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the cleaned dataset
df = pd.read_csv(r"C:\Users\priyadharshini\Desktop\Priyaa\dataset\cleaned_reviews.csv", encoding="latin1") 

#bar plot for sentiment distribution
def plot_sentiment_distribution(df):
    """Enhanced visualization of sentiment distribution."""
    plt.figure(figsize=(8, 5))
    sns.barplot(
        x=df['Sentiment'].value_counts().index, 
        y=df['Sentiment'].value_counts().values, 
        palette='viridis'
    )
    
    plt.title('Sentiment Distribution', fontsize=14, fontweight='bold')
    plt.xlabel('Sentiment', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    
    for index, value in enumerate(df['Sentiment'].value_counts().values):
        plt.text(index, value + 5, str(value), ha='center', fontsize=12, fontweight='bold')
  
    sns.despine()
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout()    
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

plot_sentiment_distribution(df)
```
